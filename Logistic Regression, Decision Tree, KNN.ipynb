#**Preprocessing**

```python
data = df.drop(columns=["Nomor peserta","Nomor keluarga", "Bobot", "Provinsi FKTP",
                        "Kabupaten/Kota FKTP", "Kepemilikan FKTP", "Kode dan nama diagnosis ICD 10 (3 digit)",
                        "Kode diagnosis ICD 10 (3 digit)", "Nama diagnosis berasal dari kode diagnosis FKP15",
                        "Kabupaten/Kota faskes tujuan rujukan",	"Kepemilikan faskes tujuan rujukan",	"Jenis faskes tujuan rujukan",
                        "Tipe faskes tujuan rujukan",	"Poli faskes tujuan rujukan"])
data.loc[df["Provinsi faskes tujuan rujukan"] == 98, "Provinsi faskes tujuan rujukan"] = 0
data.loc[df["Provinsi faskes tujuan rujukan"] != 98, "Provinsi faskes tujuan rujukan"] = 1
data
```

```python
data['Tanggal pulang kunjungan FKTP'] = pd.to_datetime(data['Tanggal pulang kunjungan FKTP'], errors='coerce')
data['Tanggal datang kunjungan FKTP'] = pd.to_datetime(data['Tanggal datang kunjungan FKTP'], errors='coerce')
data['Lama Kunjungan'] = data['Tanggal pulang kunjungan FKTP'] - data['Tanggal datang kunjungan FKTP']
```

```python
databla = data.drop(columns=["Tanggal datang kunjungan FKTP","Tanggal pulang kunjungan FKTP"])
databla
```

```python
df1 = databla.sort_values(by='Kode diagnosis (beragam 3-5 digit)',ascending=True).reset_index()
df1
```

```python
databla.shape
```

```python
databla.info()
```

```python
from sklearn.preprocessing import LabelEncoder
cnc=LabelEncoder()
```

```python
databla['Kelas Status Pulang Peserta']=cnc.fit_transform(databla['Kelas Status Pulang Peserta'].values)
databla['Lama Kunjungan']=databla['Lama Kunjungan'].dt.days
databla['Kode diagnosis (beragam 3-5 digit)']=cnc.fit_transform(databla['Kode diagnosis (beragam 3-5 digit)'].values)
databla['Jenis Poli FKTP']=cnc.fit_transform(databla['Jenis Poli FKTP'].values)
```

Reglog

```python
#split dataset
x = databla[['Jenis FKTP', 'Tipe FKTP',
       'Tingkat Pelayanan FKTP', 'Jenis Poli FKTP',
       'Segmen Peserta saat akses layanan FKTP',
       'Provinsi faskes tujuan rujukan', 'Jenis Kunjungan FKTP',
       'Lama Kunjungan', 'Kode diagnosis (beragam 3-5 digit)']]
y = databla['Kelas Status Pulang Peserta']

from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2, random_state = 88)
```

```python
#logistic regression
import warnings
warnings.filterwarnings('ignore')

from sklearn.linear_model import LogisticRegression
logreg = LogisticRegression(random_state = 88)
logreg.fit(x_train, y_train)
```

```python
y_predict = logreg.predict(x_test)
y_predict
```

```python
y_predict1 = logreg.predict_proba(x_test)
y_predict1
```

```python
#metode evaluasi
from sklearn.metrics import classification_report, confusion_matrix
print('\nconfustion matrix') # generate the confusion matrix
print(confusion_matrix(y_test, y_predict))

from sklearn.metrics import accuracy_score
print('\naccuracy')
print(accuracy_score(y_test, y_predict))

from sklearn.metrics import classification_report
print('\nclassification report')
print(classification_report(y_test, y_predict))
```

## **Decission Tree**

```python
#decission tree
from sklearn.tree import DecisionTreeClassifier
dt = DecisionTreeClassifier(random_state = 88)
dt.fit(x_train, y_train)
```

```python
y_predict = dt.predict(x_test)
```

```python
#metode evaluasi
from sklearn.metrics import classification_report, confusion_matrix
print('\nconfustion matrix') # generate the confusion matrix
print(confusion_matrix(y_test, y_predict))

from sklearn.metrics import accuracy_score
print('\naccuracy')
print(accuracy_score(y_test, y_predict))

from sklearn.metrics import classification_report
print('\nclassification report')
print(classification_report(y_test, y_predict))
```

KNN

```python
#K-NEARREST NEIGHBOR
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier()
knn.fit(x_train, y_train)
```

```python
y_predict = knn.predict(x_test)
```

```python
#metode evaluation
from sklearn.metrics import classification_report, confusion_matrix
print('\nconfustion matrix') # generate the confusion matrix
print(confusion_matrix(y_test, y_predict))

from sklearn.metrics import accuracy_score
print('\naccuracy')
print(accuracy_score(y_test, y_predict))

from sklearn.metrics import classification_report
print('\nclassification report')
print(classification_report(y_test, y_predict))
```
